```python
import scipy.stats as st
import numpy as np
```

***Example 1***

```python
t_table = st.t.ppf(1-0.05,24)
t_table
```

```python
t_cal = (148-150)/(5/np.sqrt(25))
t_cal
```

```python
if t_table < t_cal:
  print("Yes, Ha is correct")
else:
  print("No, Ha is not Correct")
```
