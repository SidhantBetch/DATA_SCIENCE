***Mean Absolute Deviation***

```python
sec_a = np.array([75,65,73,68,72,76])
sec_b = np.array([90,47,43,96,93,51])
no = np.array([1,2,3,4,5,6])
```

```python
# find mean
mean = np.mean(sec_a)
# find mean absolute deviation of section A
mad_a = np.sum(np.abs(sec_a-mean))/len(sec_a)
```

```python
# find mean of section B
mean = np.mean(sec_b)
# find mean absolute deviation of section B
mad_b = np.sum(np.abs(sec_b-mean))/len(sec_b)
```

```python
# comparition of MAD of section a and section b
(mad_a,mad_b)
```

```python
# comparing both section in Graph
plt.figure(figsize=(10,3))
plt.scatter(sec_a,no,label="sec A")
plt.scatter(sec_b,no,color="red",label="sec B")
plt.plot([70,70,70,70,70,70],no,c="blue",label="Mean")
plt.legend()
plt.show()
```

***Standerd Deviation***

```python
# find section A standerd deviation
std_a = np.std(sec_a)
# find section B standerd Deviation
std_b = np.std(sec_b)
```

```python
# comparing both
std_a, std_b
```

***Variance***

```python
# find section A Variance
var_a = np.var(sec_a)
# find section B variance
var_b = np.var(sec_b)
```

```python
# comparing both
var_a, var_b
```
