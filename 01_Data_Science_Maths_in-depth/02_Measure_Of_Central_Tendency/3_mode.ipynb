```python
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv("Titanic-Dataset.csv")
```

```python
dataset.head(3)
```

```python
# Help to find null value
dataset.isnull().sum()
```

```python
# in hull value store mean value
dataset["Fare"].fillna(dataset["Fare"].mean(),inplace=True)
```

```python
# find the mode
mo = dataset["Fare"].mode()[0]
```

```python
# find the median
md = np.median(dataset["Fear"])
md
```

```python
# other methord to find median
dataset["Fare"].median()
```

```python
# find mean
mn = dataset["Fare"].mean()
mn
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.histplot(x="Fare",data=dataset, bins=[i for i in range(0,81,10)])
plt.plot([mn for i in range(0,390)],[i for i in range(0,390)],c="red",label="mean")
plt.plot([md for i in range(0,390)],[i for i in range(0,390)],c="yellow",label="median")
plt.plot([mo for i in range(0,390)],[i for i in range(0,390)],c="blue",label="mode")
plt.legend()
plt.show()
```
