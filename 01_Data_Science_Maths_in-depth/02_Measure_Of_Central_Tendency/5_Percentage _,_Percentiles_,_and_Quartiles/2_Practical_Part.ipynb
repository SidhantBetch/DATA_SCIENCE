```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv("Titanic-Dataset.csv")
```

```python
dataset.head(3)
```

```python
# remove null value from numerical data
dataset['Age'].fillna(dataset['Age'].mean(),inplace=True)
```

```python
dataset['Age']
```

***Percentile***

```python
# find percentile
np.percentile(dataset['Age'],50)
```

***Outlier***

```python
# show IQR table which help to find outlier
dataset.describe()
```

```python
# show outlier in graph
sns.boxplot(x="Age",data=dataset)
plt.show()
```
