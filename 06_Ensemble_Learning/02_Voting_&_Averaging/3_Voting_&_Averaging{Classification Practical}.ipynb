```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("package.csv")
```

```python
dataset.head(2)
```

split data

```python
x = dataset.iloc[:,:-1]
y = dataset["package"]
```

Train Test Split data

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.25,random_state=42)
```

# ***Voting And Averaging(Classification)***

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from sklearn.linear_model import LinearRegression
```

***Train ml using LinearRegression***

```python
lr = LinearRegression()
lr.fit(x_train,y_train)
lr.score(x_train,y_train)*100, lr.score(x_test,x_test)
```

***Train ml using SVR***

```python
sv = SVR()
sv.fit(x_train,y_train)
sv.score(x_train,y_train)*100, sv.score(x_test,x_test)
```

***Train ml using DecisionTreeRegressor***

```python
dt = DecisionTreeRegressor()
dt.fit(x_train,y_train)
dt.score(x_train,y_train)*100, dt.score(x_test,x_test)
```

Voting Regressor

```python
from sklearn.ensemble import VotingRegressor
```

```python
li = [("lr1",LinearRegression()),("dt1",DecisionTreeRegressor()),("sv1",SVR())]
```

```python
vc = VotingRegressor(li, weights=[30,5,30])
vc.fit(x_train,y_train)
```

```python
vc.score(x_train,y_train)*100, vc.score(x_test,y_test)*100
```

check how predict work

```python
prd = {"dt":dt.predict(x_test),"svm":sv.predict(x_test),"lr":lr.predict(x_test),"vc":vc.predict(x_test)}
```
