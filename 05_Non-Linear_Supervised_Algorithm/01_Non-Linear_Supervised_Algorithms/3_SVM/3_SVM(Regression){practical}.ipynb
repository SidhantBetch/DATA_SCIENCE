```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from mlxtend.plotting import plot_decision_regions
```

```python
dataset = pd.read_csv("package.csv")
```

```python
dataset.head(3)
```

```python
plt.figure(figsize=(3,2))
sns.scatterplot(x="cgpa", y="package", data=dataset)
plt.show()
```

split data

```python
x = dataset[["cgpa"]]
y = dataset["package"]
```

# split data into train and test

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,train_size=0.25,random_state=22)
```

# ***SVM***

***SVM (Regression)***

```python
from sklearn.svm import SVR
```

kernel = Linear

```python
sv = SVR(kernel="linear")
sv.fit(x_train, y_train)
```

```python
print("(Test Score=",sv.score(x_test,y_test)*100,", Train Score =",sv.score(x_train, y_train)*100,")")
```

```python
sns.scatterplot(x="cgpa",y="package",data = dataset)
plt.plot(dataset["cgpa"],sv.predict(x),c="red")
plt.show()
```

Kernal = Polynomial

```python
sv2 = SVR(kernel="poly")
sv2.fit(x_train, y_train)
```

```python
print("(Test Score=",sv2.score(x_test,y_test)*100,", Train Score =",sv2.score(x_train, y_train)*100,")")
```

```python
sns.scatterplot(x="cgpa",y="package",data = dataset)
plt.plot(dataset["cgpa"],sv2.predict(x),c="red")
plt.show()
```

Kernel = rbf

```python
sv3 = SVR(kernel="rbf")
sv3.fit(x_train, y_train)
```

```python
print("(Test Score=",sv3.score(x_test,y_test)*100,", Train Score =",sv3.score(x_train, y_train)*100,")")
```

```python
sns.scatterplot(x="cgpa",y="package",data = dataset)
plt.plot(dataset["cgpa"],sv3.predict(x),c="red")
plt.show()
```

Kernel = sigmold

```python
sv4 = SVR(kernel="sigmoid")
sv4.fit(x_train, y_train)
```

```python
print("(Test Score=",sv4.score(x_test,y_test)*100,", Train Score =",sv4.score(x_train, y_train)*100,")")
```

```python
sns.scatterplot(x="cgpa",y="package",data = dataset)
plt.plot(dataset["cgpa"],sv4.predict(x),c="red")
plt.show()
```
