```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from mlxtend.plotting import plot_decision_regions
```

```python
dataset = pd.read_csv("placement-dataset.csv")

dataset.drop(columns=["Unnamed: 0"],inplace=True)
```

```python
dataset.head(3)
```

```python
plt.figure(figsize=(3,2))
sns.scatterplot(x="cgpa", y="iq", data=dataset, hue="placement")
plt.show()
```

split data

```python
x = dataset.iloc[:,:-1]
y = dataset["placement"]
```

# split data into train and test

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,train_size=0.2,random_state=42)
```

# ***SVM***

***SVM (classification)***

```python
from sklearn.svm import SVC
```

Linear

```python
sv = SVC(kernel="linear")
sv.fit(x_train, y_train)
```

```python
print("(Test Score=",sv.score(x_test,y_test)*100,", Train Score =",sv.score(x_train, y_train)*100,")")
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=sv)
plt.show()
```

Kernal = Polynomial

```python
sv2 = SVC(kernel="poly")
sv2.fit(x_train, y_train)
```

```python
print("(Test Score=",sv2.score(x_test,y_test)*100,", Train Score =",sv2.score(x_train, y_train)*100,")")
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=sv2)
plt.show()
```

Kernel = rbf

```python
sv3 = SVC(kernel="rbf")
sv3.fit(x_train, y_train)
```

```python
print("(Test Score=",sv3.score(x_test,y_test)*100,", Train Score =",sv3.score(x_train, y_train)*100,")")
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=sv3)
plt.show()
```

Kernel = sigmold

```python
sv4 = SVC(kernel="sigmoid")
sv4.fit(x_train, y_train)
```

```python
print("(Test Score=",sv4.score(x_test,y_test)*100,", Train Score =",sv4.score(x_train, y_train)*100,")")
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=sv4)
plt.show()
```
