```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("iris.csv")

dataset.drop(columns=["Unnamed: 0","Species"],inplace=True)
```

```python
dataset.head(3)
```

```python
plt.figure(figsize=(3,3))
sns.pairplot(data=dataset)
plt.show()
```

# *CLUSTRING*

***K-MEAN CLUSTRING***

```python
from sklearn.cluster import KMeans
```

```python
wcss = []

for i in range(2,21):
  km = KMeans(n_clusters=i, init='k-means++')
  km.fit(dataset)
  wcss.append(km.inertia_)
```

```python
plt.figure(figsize=(10,5))
plt.plot([i for i in range(2,21)],wcss,marker = "o")
plt.xlabel("no of Cluster")
plt.xticks([i for i in range(2,21)])
plt.ylabel("wcss")
plt.grid(axis='x')
plt.show()
```

```python
kmn = KMeans(n_clusters=3)
dataset["Predict"] = kmn.fit_predict(dataset)
```

***SILHOUETTE SCORE***

```python
from sklearn.metrics import silhouette_score
```

```python
# silhouette_score(dataset,labels = kmn.labels_)
```

```python
ss = []
no_c = [j for j in range(2,21)]
for i in range(2,21):
  km1 = KMeans(n_clusters=i)
  km1.fit(dataset)
  ss.append(silhouette_score(dataset,km1.labels_))
```

```python
plt.plot(no_c, ss)
plt.xlabel("No. of cluster")
plt.ylabel("silhouette_Score")
plt.xticks(no_c)
plt.grid(axis='x')
plt.show()
```
