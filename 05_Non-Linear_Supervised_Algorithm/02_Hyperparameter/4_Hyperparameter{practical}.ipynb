```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv(r"/content/Social_Network_Ads.csv")

dataset.drop(columns=["User ID","Gender"],inplace=True)
```

```python
dataset.head(3)
```

split data

```python
x = dataset.iloc[:,:-1]
y = dataset["Purchased"]
```

# split data into train and test

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,train_size=0.25,random_state=22)
```

***DecisionTreeRegressor***

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dt = DecisionTreeRegressor(max_depth=2)
dt.fit(x_train, y_train)
```

```python
print(f"Test Score = {dt.score(x_test, y_test)*100:.2f}%, Train Score = {dt.score(x_train, y_train)*100:.2f}%")
```

# ***Hyperparameter***

Gridsearchcv

```python
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
```

```python
df = {
    "criterion": ["squared_error","friedman_mse","absolute_error","poisson"],
    "splitter": ["best","random"],
    "max_depth":[i for i in range(2,20)]
}
```

```python
gd = GridSearchCV(DecisionTreeRegressor(), param_grid=df, cv=5)
gd.fit(x_train, y_train)
```

```python
gd.best_params_
```

after using hyperparameter

```python
dt1 = DecisionTreeRegressor(criterion="absolute_error",max_depth=2,splitter="best")
dt1.fit(x_train, y_train)
```

```python
print(f"Test Score = {dt1.score(x_test, y_test)*100:.2f}%, Train Score = {dt1.score(x_train, y_train)*100:.2f}%")
```

RandomSearchcv

```python
rd = RandomizedSearchCV(
    DecisionTreeRegressor(),   # ✅ object, not class
    param_distributions=df,
    n_iter=20,
    cv=5
)

rd.fit(x_train, y_train)
```

```python
rd.best_params_
```

After hyperparameter

```python
dt2 = DecisionTreeRegressor(criterion="friedman_mse",max_depth=2,splitter="best")
dt2.fit(x_train,y_train)
```

```python
print(f"Test Score = {dt2.score(x_test, y_test)*100:.2f}%, Train Score = {dt2.score(x_train, y_train)*100:.2f}%")
```
