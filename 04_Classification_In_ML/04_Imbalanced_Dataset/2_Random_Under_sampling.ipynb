```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("Social_Network_Ads.csv")

dataset.drop(columns=["User ID","Gender"],inplace=True)
```

```python
dataset.head(3)
```

value_counts help to find total  no. of each data

```python
dataset["Purchased"].value_counts()
```

whithout Imbalancing data

```python
x = dataset.iloc[:,:-1]
y = dataset["Purchased"]
```

Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,train_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
lr.fit(x_train, y_train)
```

```python
lr.score(x_test,y_test)*100
```

ML predict wrong value because our data is Bayes

```python
lr.predict([[45,26000]])
```

# ***Import imblearn for imbalanced dataset***

## Random Under Sampling

```python
from imblearn.under_sampling import RandomUnderSampler
```

```python
ru = RandomUnderSampler()
ru_x, ru_y = ru.fit_resample(x,y)
```

check the data is imbalanced or not

```python
ru_y.value_counts()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train1, x_test1, y_train1, y_test1 = train_test_split(ru_x,ru_y,train_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr1 = LogisticRegression()
lr1.fit(x_train1, y_train1)
```

```python
lr1.score(x_test1,y_test1)*100
```

```python
lr1.predict([[45,26000]])
```
