```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("TrainDataset.csv")
dataset.head(3)
```

```python
dataset.isnull().sum()
```

***choose column which only countable data repidated or there was not null value***

```python
en_data = dataset[["Gender","Married"]]
pd.get_dummies(en_data)
```

# ***One Hot Encoding***

```python
# sklearn is learn data and convert categorical data into numerical data
from sklearn.preprocessing import OneHotEncoder
```

```python
# convert simple data into onehotEncoder
ohe = OneHotEncoder()
ohe.fit_transform(en_data)
```

```python
# store oneHotEncoding data into array formate
ar = ohe.fit_transform(en_data).toarray()
ar
```

```python
# store oneHotEncoder data into data frame formate
pd.DataFrame(ar, columns=[
    "Gender_Female",
    "Gender_Male",
    "Married_No",
    "Married_Yes"
])
```

# ***Remove one column for reduce no. of column***

```python
# remove first data from oneHot
ohe2 = OneHotEncoder(drop="first")
arr = ohe2.fit_transform(en_data).toarray()
arr
```

```python
pd.DataFrame(arr, columns=[
    "Gender_Male",
    "Married_Yes"
])
```
