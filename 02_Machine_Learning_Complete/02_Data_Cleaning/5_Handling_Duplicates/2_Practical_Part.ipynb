# ***Duplicate Handling***

```python
data = {"name":["a","b","c","d","a","c"],"eng":[8,7,5,8,8,4],"hindi":[2,3,4,5,2,6]}
```

```python
df = pd.DataFrame(data)
df
```

```python
df_dup = df.duplicated()
df_dup
```

# ***Duplicate Remove***

```python
df.drop_duplicates(inplace = True)
df
```
