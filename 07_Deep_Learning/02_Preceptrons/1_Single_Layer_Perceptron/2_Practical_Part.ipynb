```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from mlxtend.plotting import plot_decision_regions
```

```python
dataset = pd.read_csv(r"/content/placement-dataset.csv")

dataset.drop(columns=["Unnamed: 0"],inplace=True)
```

```python
dataset.head(2)
```

```python
plt.figure(figsize=(4,3))
sns.scatterplot(x="cgpa",y="iq",data = dataset, hue="placement")
plt.show()
```

split data

```python
x = dataset.iloc[:,:-1]
y = dataset["placement"]
```

Train Test Split data

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.25,random_state=10)
```

# ***Single Layer Precepton***

```python
from sklearn.linear_model import Perceptron
```

```python
pr = Perceptron()
pr.fit(x_train,y_train)
```

```python
pr.score(x_train,y_train)*100, pr.score(x_test,y_test)*100
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=pr)
plt.show()
```
