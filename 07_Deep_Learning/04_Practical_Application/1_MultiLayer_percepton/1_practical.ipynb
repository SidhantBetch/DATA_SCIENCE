```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
```

```python
dataset = pd.read_csv("Churn_Modelling.csv")
```

```python
dataset.head(2)
```

Insure data has not null value

```python
dataset.isnull().sum()
```

```python
input_data = dataset.iloc[:,:-1]
output_data = dataset["Exited"]
```

Scaling

```python
ss = StandardScaler()
input_data = pd.DataFrame(ss.fit_transform(input_data),columns=input_data.columns)
```

Train Test Split data

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(input_data,output_data,train_size=0.25,random_state=10)
```

```python
x_train.shape
```

# ***MultiLayer Percepton***

```python
import tensorflow
```

```python
from keras.layers import Dense
from keras.models import Sequential
```

```python
ann = Sequential()
```

```python
ann.add(Dense(6,input_dim = 8,activation="relu"))
ann.add(Dense(4,activation="relu"))
ann.add(Dense(2,activation="relu"))
ann.add(Dense(1,activation="sigmoid"))
```

```python
ann.compile(optimizer = "adam", loss = "binary_crossentropy",metrics=['accuracy'] )
```

```python
ann.fit(x_train,y_train, batch_size=100, epochs = 50)
```

check accuracy

```python
prd = ann.predict(x_test)
prd_data = []
for i in prd:
  if i[0] > 0.5 :
    prd_data.append(1)
  else:
    prd_data.append(0)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,prd_data)*100
```

```python
prd1 = ann.predict(x_train)
prd_data1 = []
for i in prd1:
  if i[0] > 0.5 :
    prd_data1.append(1)
  else:
    prd_data1.append(0)
```

```python
accuracy_score(y_train,prd_data1)*100
```
