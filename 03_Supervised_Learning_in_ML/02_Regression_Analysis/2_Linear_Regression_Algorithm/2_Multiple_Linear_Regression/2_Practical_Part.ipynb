```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv("employee_performance.csv")
```

```python
dataset.head(3)
```

```python
dataset.isnull().sum()
```

```python
dataset.shape
```

drop all unnessesary column and remove null value

```python
dataset.drop(["Performance_Score","Employee_ID","Gender","Department"], axis=1, inplace=True)
dataset["Experience_Years"] = dataset["Experience_Years"].fillna(dataset["Experience_Years"].mean())
dataset.isnull().sum()
```

```python
plt.figure(figsize=(4,3))
sns.pairplot(data=dataset)
plt.show()
```

```python
sns.heatmap(data=dataset.corr(), annot=True)
plt.show()
```

# ***Multiple Linear Regression Algorithm***

```python
x= dataset.iloc[:,:-1]
y = dataset["Salary"]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,train_size=0.25)
```

```python
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
lr.score(x_test, y_test)*100
```

```python
lr.predict(x_test)
```
