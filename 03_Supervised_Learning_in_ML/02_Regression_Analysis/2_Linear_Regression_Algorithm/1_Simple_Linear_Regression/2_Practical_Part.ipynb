# ***Simple Linear Regression Algorithm***

```python
dataset.isnull().sum()
```

```python
x = dataset["cgpa"] # in sklearn, dataset is always in multi dimensional
x.ndim # show the dimension so we use double squre breaket
x = dataset[["cgpa"]]
x.ndim
```

```python
y = dataset["package"] # in skelearn, output data will be single dimestional
```

Train-test-split in dataset

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=25)
```

import linear regression

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
lr.fit(x_train,y_train)
```

check our model

```python
lr.predict([[6.89]])
```

find the our acuracy score

```python
lr.score(x_test,y_test)*100
```

show prediction line in graph using matplotlib and seaborn

```python
y_pred = lr.predict(x)
```

```python
plt.figure(figsize=(4,3))
sns.scatterplot(x="cgpa",y="package", data=dataset)
plt.plot(dataset["cgpa"],y_pred, c="red")
plt.legend(["org data", "predict Line"])
plt.show()
```
